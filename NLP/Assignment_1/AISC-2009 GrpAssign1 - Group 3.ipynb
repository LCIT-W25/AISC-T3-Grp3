# Que-1 How many records have a date that is expressed without using alphabets?

```python
# Regex pattern to match different date formats
pattern = r"\d{1,2}[-/]\d{1,2}[-/]\d{2,4}|\d{2,4}[-/]\d{1,2}[-/]\d{1,2}|\d{4}-\d{2}, \d{1,2}"
contains_match = df['Data'].str.contains(pattern, na=False)

print(df[contains_match])
print(f"Number of matching rows: {df[contains_match].shape[0]}")
```

# Que-2 How many records have a word starting with the letter “w”?

```python
# Regex pattern to match words starting with "w" or "W"
pattern = r'^[wW]'
matches = df['Data'].str.contains(pattern, na=False)

print(f"Number of records with a word starting with 'w': {matches.sum()}")
```

# Que-3 How many records make a word that starts with an alphabet and is not a URL?

```python
# Regex pattern to match words that start with an alphabet (case-insensitive)
# and ensure they are not part of a URL (e.g., excluding words starting with "http", "https", "www")
pattern = r'(?!http|https|www)[a-zA-Z]+'
matches = df['Data'].str.contains(pattern, case=False, na=False)

print(f"Number of records with a word starting with an alphabet and not a URL: {matches.sum()}")
```

# Que-4 How many tweets contain one of these emojis :), :D, ;), :P?

```python
# Regex pattern to match the emojis ":)", ":D", ";)", ":P"
pattern = r'(?:\)|:D|;\)|:P)'
matches = df['Data'].str.contains(pattern, na=False, regex=True)

print(f"Number of tweets containing one of the specified emojis: {matches.sum()}")
```

# Que-5 How many records contain a decimal number?

```python
# Regex pattern to match decimal numbers (e.g., 12.34, 0.5, 3.14)
pattern = r'\d+\.\d+'
matches = df['Data'].str.contains(pattern, na=False, regex=True)

print(f"Number of records containing a decimal number: {matches.sum()}")
```

# Que-6 What is the total number of ip addresses across all the records?

```python
# Updated regex pattern to strictly match IPv4 addresses (0-255 in each octet)
pattern = r'(?:25[0-5]|2[0-4][0-9]|[01]?[0-9][0-9]?)\.(?:25[0-5]|2[0-4][0-9]|[01]?[0-9][0-9]?)\.(?:25[0-5]|2[0-4][0-9]|[01]?[0-9][0-9]?)\.(?:25[0-5]|2[0-4][0-9]|[01]?[0-9][0-9]?)'
matches = df['Data'].fillna('').str.findall(pattern)

# Display the filtered records
print("Records containing valid IP addresses:", matches.apply(len).sum())
```

# Que-7 How many records have a new line?

```python
# Regex pattern to match a newline character
pattern = r'\n'
matches = df['Data'].str.contains(pattern, na=False)

print(f"Number of records containing a newline: {matches.sum()}")
```

# Que-8 What is the total number of hashtags across all these tweets?

# Que-9 What is the code to substitute all non-alphanumeric characters with a new line?

```python
# Regex pattern to match all non-alphanumeric characters
pattern = r'[^a-zA-Z0-9\s]'

# Substitute all non-alphanumeric characters with a newline
df['Data'] = df['Data'].str.replace(pattern, '\n', regex=True)

# Display the modified DataFrame
print(df)
```

# Que-1 Use porter stemmer to run stemming. Count the number of unique words/tokens before and after stemming

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import re
import nltk
from nltk.stem import WordNetLemmatizer
from collections import Counter
```

```python
# Download required NLTK data
# nltk.download('punkt')
# nltk.download('stopwords')
# nltk.download('wordnet')
# nltk.download('omw-1.4')  # For WordNet's extended morphological database
```

```python
# Initialize Porter Stemmer
stemmer = PorterStemmer()

# Preprocessing: Tokenize and clean text
def preprocess_text(text):
    # Remove non-alphanumeric characters
    text = re.sub(r"[^\w\s]", "", text)
    # Tokenize text
    tokens = word_tokenize(text.lower())
    # Remove stopwords
    tokens = [word for word in tokens if word not in stopwords.words('english')]
    return tokens

# Apply preprocessing
df.loc[:, "Tokens"] = df["Data"].apply(preprocess_text)

# Count unique words/tokens before stemming
all_tokens = [token for tokens in df["Tokens"] for token in tokens]
unique_tokens_before = set(all_tokens)

# Apply stemming
df.loc[:, "Stemmed_Tokens"] = df["Tokens"].apply(
    lambda tokens: [stemmer.stem(token) for token in tokens]
)

# Count unique words/tokens after stemming
all_stemmed_tokens = [token for tokens in df["Stemmed_Tokens"] for token in tokens]
unique_tokens_after = set(all_stemmed_tokens)

# Display counts
print(f"Number of unique words before stemming: {len(unique_tokens_before)}")
print(f"Number of unique words after stemming: {len(unique_tokens_after)}")
```

# Que-2 Perform lemmatization using NLTK lemmatizer. Count the number of unique words/tokens before and after lemmatization

```python
# Initialize the lemmatizer
lemmatizer = WordNetLemmatizer()

# Function to preprocess text (e.g., tokenization)
def preprocess_text(text):
    tokens = word_tokenize(text)  # Tokenize the text
    return tokens

# Apply tokenization to the 'Data' column
df.loc[:, "Tokens"] = df["Data"].apply(preprocess_text)

# Count unique tokens before lemmatization
unique_tokens_before = set(token for tokens in df['Tokens'] for token in tokens)
print(f"Number of unique tokens before lemmatization: {len(unique_tokens_before)}")

# Function to perform lemmatization
def lemmatize_tokens(tokens):
    return [lemmatizer.lemmatize(token) for token in tokens]

# Apply lemmatization to the tokens
df.loc[:, 'Lemmatized_Tokens'] = df['Tokens'].apply(lemmatize_tokens)

# Count unique tokens after lemmatization
unique_tokens_after = set(token for tokens in df['Lemmatized_Tokens'] for token in tokens)
print(f"Number of unique tokens after lemmatization: {len(unique_tokens_after)}")
```

# Que-3 Compare the change in word frequencies from stemming and lemmatization. Which are the top 10 words after stemming/lemmatization

```python
# Calculate word frequencies after stemming
stemmed_words = [word for tokens in df['Stemmed_Tokens'] for word in tokens]
stemmed_word_freq = Counter(stemmed_words)

# Calculate word frequencies after lemmatization
lemmatized_words = [word for tokens in df['Lemmatized_Tokens'] for word in tokens]
lemmatized_word_freq = Counter(lemmatized_words)

# Get the top 10 most frequent words for stemming and lemmatization
top_stemmed_words = stemmed_word_freq.most_common(10)
top_lemmatized_words = lemmatized_word_freq.most_common(10)

# Display the results
print("Top 10 words after stemming:")
for word, freq in top_stemmed_words:
    print(f"{word}: {freq}")

print("\nTop 10 words after lemmatization:")
for word, freq in top_lemmatized_words:
    print(f"{word}: {freq}")
```

# Que-4 What is the change in word frequencies if normalization is done after stop word removal.

```python
# Load stop words
stop_words = set(stopwords.words('english'))

# Function to remove stop words
def remove_stop_words(tokens):
    return [word for word in tokens if word.lower() not in stop_words]

# Remove stop words from the tokens
df.loc[:, 'Tokens_No_Stopwords'] = df['Tokens'].apply(remove_stop_words)

stemmer = PorterStemmer()

# Define the stemming function
def stem_tokens(tokens):
    return [stemmer.stem(token) for token in tokens]

# Apply stemming after stop word removal
df.loc[:, 'Stemmed_No_Stopwords'] = df['Tokens_No_Stopwords'].apply(stem_tokens)

# Initialize lemmatizer
lemmatizer = WordNetLemmatizer()

# Function to lemmatize tokens
def lemmatize_tokens(tokens):
    return [lemmatizer.lemmatize(token) for token in tokens]

# Apply lemmatization after stop word removal
df.loc[:, 'Lemmatized_No_Stopwords'] = df['Tokens_No_Stopwords'].apply(lemmatize_tokens)

# Calculate word frequencies after stop word removal
tokens_no_stopwords = [word for tokens in df['Tokens_No_Stopwords'] for word in tokens]
freq_no_stopwords = Counter(tokens_no_stopwords)

# Calculate word frequencies after stemming
stemmed_no_stopwords = [word for tokens in df['Stemmed_No_Stopwords'] for word in tokens]
freq_stemmed_no_stopwords = Counter(stemmed_no_stopwords)

# Calculate word frequencies after lemmatization
lemmatized_no_stopwords = [word for tokens in df['Lemmatized_No_Stopwords'] for word in tokens]
freq_lemmatized_no_stopwords = Counter(lemmatized_no_stopwords)

# Compare the frequencies before and after normalization
top_no_stopwords = freq_no_stopwords.most_common(10)
top_stemmed_no_stopwords = freq_stemmed_no_stopwords.most_common(10)
top_lemmatized_no_stopwords = freq_lemmatized_no_stopwords.most_common(10)

# Display the results
print("Top 10 words after stop word removal:")
for word, freq in top_no_stopwords:
    print(f"{word}: {freq}")

print("\nTop 10 words after stemming (post stop word removal):")
for word, freq in top_stemmed_no_stopwords:
    print(f"{word}: {freq}")

print("\nTop 10 words after lemmatization (post stop word removal):")
for word, freq in top_lemmatized_no_stopwords:
    print(f"{word}: {freq}")
```
