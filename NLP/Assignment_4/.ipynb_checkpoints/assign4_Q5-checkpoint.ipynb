```python
# Data Preprocessing & Feature Engineering
```

```python
import pandas as pd

# Taken less data due to less memory and computation power of laptop

df = pd.read_csv("Sentiment_Data.csv", encoding='latin-1')
df_sample = df.sample(n=1000,random_state=42).reset_index(drop=True)
print(df_sample.head())
print("\n Sample size: ", df_sample.shape)

df_sample['Sentiment'].unique()
sentiment_cnt = df_sample['Sentiment'].value_counts()
print("\n Sentiment Count: ", sentiment_cnt)

# Check for missing values in the dataset
missing_values = df_sample.isnull().sum()
print("\n Missing values in each column:")
print(missing_values)
```

```python
# Cleaning Tweets
import re

# Complete clean_text function for BERTweet
def clean_text(text):
    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"@\w+", "", text)
    text = re.sub(r"#\w+", "", text)
    text = text.encode('ascii', 'ignore').decode('ascii')
    text = text.lower()
    text = re.sub(r"\s+", " ", text).strip()
    return text

# Apply cleaning function to the dataset
df_sample['Cleaned_Text'] = df_sample['Tweet'].apply(clean_text)

# Display first 5 cleaned tweets
print(df_sample[['Tweet', 'Cleaned_Text']].head())

print("\n\n Original Tweet: ",  df_sample['Tweet'][2])
print("\n Cleaned Tweet: ",df_sample['Cleaned_Text'][2])
```

```python
from transformers import AutoTokenizer
import torch

# Tokenization and Padding using DistilBERT
# Using the DistilBERT tokenizer (lighter and optimized for less computation)
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")

# Tokenize the cleaned text
# We use a batch size for tokenization to avoid memory issues during processing
def tokenize_data(texts):
    return tokenizer(texts, padding=True, truncation=True, return_tensors="pt", max_length=64)

# Sample input (cleaned texts from the dataset)
cleaned_texts = df_sample['Cleaned_Text'].tolist()

# Tokenizing the dataset (using only a subset to keep memory usage lower)
batch_size = 200  
input_ids = []
attention_masks = []

for i in range(0, len(cleaned_texts), batch_size):
    batch = cleaned_texts[i:i+batch_size]
    tokenized = tokenize_data(batch)
    input_ids.append(tokenized['input_ids'])
    attention_masks.append(tokenized['attention_mask'])

# Concatenate the input IDs and attention masks from all batches
input_ids = torch.cat(input_ids, dim=0)
attention_masks = torch.cat(attention_masks, dim=0)

# Check tokenization output for the first sample
print("\nFirst Tokenized Input IDs:", input_ids[0])
print("\nFirst Attention Mask:", attention_masks[0])

# Convert labels to numerical values (for sentiment classification)
label_mapping = {
    'Strong_Pos': 0,  # 0 for Strong Positive
    'Mild_Pos': 1,    # 1 for Mild Positive
    'Neutral': 2,     # 2 for Neutral
    'Mild_Neg': 3,    # 3 for Mild Negative
    'Strong_Neg': 4   # 4 for Strong Negative
}

# Map the sentiment labels to integers
df_sample['Label'] = df_sample['Sentiment'].map(label_mapping)

# Display first 5 labels to confirm correct conversion
print("\nFirst 5 Labels: ", df_sample.head())
```

```python
from transformers import AutoModel
import torch
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Load the pre-trained DistilBERT model
model = AutoModel.from_pretrained("distilbert-base-uncased")

# Function to get embeddings from the model
def get_embeddings(input_ids, attention_masks):
    with torch.no_grad():
        outputs = model(input_ids, attention_mask=attention_masks)
    # Extract the [CLS] token embeddings (usually the first token in the sequence)
    embeddings = outputs.last_hidden_state[:, 0, :]  # (batch_size, hidden_size)
    return embeddings

# Get embeddings for the tokenized inputs
embeddings = get_embeddings(input_ids, attention_masks)

# Convert embeddings to numpy array for t-SNE (make sure embeddings are of shape (batch_size, embedding_dim))
embeddings = embeddings.numpy()

# Check the shape of embeddings
print("Embeddings Shape:", embeddings.shape)  # This should be (num_samples, embedding_dim)

# Apply t-SNE for dimensionality reduction to 2D
tsne = TSNE(n_components=2, random_state=42, perplexity=30, n_iter=300)
reduced_embeddings = tsne.fit_transform(embeddings)

# Visualize the reduced embeddings with labels
plt.figure(figsize=(8, 6))
plt.scatter(reduced_embeddings[:, 0], reduced_embeddings[:, 1], c=df_sample['Label'], cmap='viridis', s=50)
plt.colorbar(label='Sentiment Class')
plt.title('Class Separability using t-SNE')
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.show()
```

```python
from transformers import DistilBertModel
import torch
import torch.nn as nn

# Load pre-trained DistilBERT model
model = DistilBertModel.from_pretrained("distilbert-base-uncased")

# Define the Embedding Layer using the pre-trained DistilBERT model
class DistilBertEmbeddingLayer(nn.Module):
    def __init__(self, model):
        super(DistilBertEmbeddingLayer, self).__init__()
        self.distilbert = model
        self.dropout = nn.Dropout(0.1)  # Optional dropout layer after embedding

    def forward(self, input_ids, attention_mask):
        # print("here::::::::::")
        # Get the embeddings (last_hidden_state) from DistilBERT
        outputs = self.distilbert(input_ids, attention_mask=attention_mask)
        embeddings = outputs.last_hidden_state  # shape: [batch_size, seq_length, hidden_size]

        # print(f"Embeddings Shape: {embeddings.shape}")  # Should be [batch_size, seq_length, hidden_size]
        # Optional: Apply dropout to embeddings for regularization
        embeddings = self.dropout(embeddings)

        return embeddings

# Initialize the embedding layer
embedding_layer = DistilBertEmbeddingLayer(model)

# Test embedding layer with sample data (using the previously tokenized data)
sample_embeddings = embedding_layer(input_ids[:2], attention_masks[:2])  # Using the first 2 samples for testing
print(f"Sample Embeddings Shape: {sample_embeddings.shape}")  # Should be [batch_size, seq_length, hidden_size]

print(model)
```

```python
import torch
import time
import torch.nn as nn
import torch.optim as optim
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, auc, classification_report
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import label_binarize
from torch.utils.data import TensorDataset, DataLoader

# Define device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Define loss and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.AdamW(model.parameters(), lr=2e-5, weight_decay=1e-4)

# Training parameters
num_epochs = 3
batch_size = 16

# Create dataset and dataloader
train_dataset = TensorDataset(train_input_ids, train_attention_masks, train_labels)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

# Function to evaluate the model
def evaluate_model(model, train_loader, device, num_classes=3):
    model.eval()
    all_preds, all_labels, all_probs = [], [], []
    
    with torch.no_grad():
        for batch in train_loader:
            input_ids, attention_masks, labels = [x.to(device) for x in batch]
            outputs = model(input_ids, attention_mask=attention_masks)
            
            logits = outputs.last_hidden_state[:, 0, :]  # CLS token representation
            
            preds = torch.argmax(logits, dim=1)
            probs = torch.softmax(logits, dim=1)
            
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
            all_probs.extend(probs.cpu().numpy())

    # Compute metrics
    accuracy = accuracy_score(all_labels, all_preds)
    cm = confusion_matrix(all_labels, all_preds)
    
    # Compute AUC
    all_labels_bin = label_binarize(all_labels, classes=range(num_classes))
    auc_scores = {}
    for i in range(num_classes):
        fpr, tpr, _ = roc_curve(all_labels_bin[:, i], np.array(all_probs)[:, i])
        auc_scores[f'Class {i} AUC'] = auc(fpr, tpr)

    # Plot ROC curve
    plt.figure(figsize=(8, 6))
    for i in range(num_classes):
        fpr, tpr, _ = roc_curve(all_labels_bin[:, i], np.array(all_probs)[:, i])
        plt.plot(fpr, tpr, label=f'Class {i} AUC = {auc_scores[f"Class {i} AUC"]:.4f}')
    plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Training ROC Curve')
    plt.legend(loc='lower right')
    plt.show()

    class_report = classification_report(all_labels, all_preds, digits=4)
    return accuracy, cm, auc_scores, class_report

# Training loop
for epoch in range(num_epochs):
    model.train()
    total_loss = 0
    start_time = time.time()

    for batch in train_loader:
        input_ids, attention_masks, labels = [x.to(device) for x in batch]

        optimizer.zero_grad()
        outputs = model(input_ids, attention_mask=attention_masks)

        logits = outputs.last_hidden_state[:, 0, :]  # CLS token representation
        
        loss = criterion(logits, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

    avg_loss = total_loss / len(train_loader)
    epoch_train_time = time.time() - start_time

    print(f"Epoch {epoch+1} - Loss: {avg_loss:.4f}")
    print(f"Epoch {epoch+1} - Train Time: {epoch_train_time:.2f} seconds")

    # Evaluate model
    accuracy, cm, auc_scores, class_report = evaluate_model(model, train_loader, device, num_classes=3)

    print(f"Epoch {epoch+1} - Train Accuracy: {accuracy:.4f}")
    print(f"Epoch {epoch+1} - Train Confusion Matrix:\n{cm}")
    for class_label, auc_score in auc_scores.items():
        print(f"{class_label}: {auc_score:.4f}")
    print(f"Epoch {epoch+1} - Train Classification Report:\n{class_report}")

print("Model Training Completed!")
```

```python
import torch
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, roc_curve, auc, classification_report
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_auc_score

# Evaluation function
def evaluate_model(model, val_loader, device, num_classes=3):  # num_classes is for multiclass classification
    model.eval()  # Set model to evaluation mode
    
    all_preds = []
    all_labels = []
    all_probs = []  # To store predicted probabilities for AUC
    
    with torch.no_grad():  # No gradients are needed for evaluation
        for batch in val_loader:
            input_ids, attention_masks, labels = [x.to(device) for x in batch]
            
            # Forward pass
            outputs = model(input_ids, attention_mask=attention_masks)
            
            # Extract logits correctly
            if isinstance(outputs, tuple):
                logits = outputs[0]  # If it's a tuple, take the first element
            elif hasattr(outputs, "logits"):
                logits = outputs.logits  # If model returns a named attribute
            elif hasattr(outputs, "last_hidden_state"):
                logits = outputs.last_hidden_state[:, 0, :]  # Use CLS token if needed
            else:
                raise ValueError("Unexpected model output format.")
            
            # Ensure logits have the expected shape for classification
            if len(logits.shape) == 2 and logits.shape[1] > 1:  # Check if it's 2D with the class dimension
                preds = torch.argmax(logits, dim=1)  # Choose the class with the highest logit value
                probs = torch.softmax(logits, dim=1)  # Get class probabilities for AUC curve
            else:
                print(f"Unexpected logits shape: {logits.shape}")
                preds = logits  # Just in case, this should not happen if everything works
                probs = logits  # This line is to handle unexpected shapes
            
            all_preds.extend(preds.cpu().numpy())  # Move to CPU for further processing
            all_labels.extend(labels.cpu().numpy())
            all_probs.extend(probs.cpu().numpy())  # Store probabilities for AUC
    
    # Calculate metrics
    accuracy = accuracy_score(all_labels, all_preds)
    f1 = f1_score(all_labels, all_preds, average='weighted')
    cm = confusion_matrix(all_labels, all_preds)
    
    # Compute AUC for multiclass classification
    all_labels_bin = label_binarize(all_labels, classes=range(num_classes))  # Binarize for each class
    auc_scores = {}
    for i in range(num_classes):
        fpr, tpr, _ = roc_curve(all_labels_bin[:, i], np.array(all_probs)[:, i])
        auc_score = auc(fpr, tpr)
        auc_scores[f'Class {i} AUC'] = auc_score
        
    # Plot the ROC curve for each class
    plt.figure(figsize=(8, 6))
    for i in range(num_classes):
        fpr, tpr, _ = roc_curve(all_labels_bin[:, i], np.array(all_probs)[:, i])
        plt.plot(fpr, tpr, label=f'Class {i} AUC = {auc_scores[f"Class {i} AUC"]:.4f}')
    
    plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # Diagonal line for random classifier
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Multiclass ROC Curve')
    plt.legend(loc='lower right')
    plt.show()

    # Generate classification report
    class_report = classification_report(all_labels, all_preds, digits=4)

    return accuracy, f1, cm, auc_scores, class_report

# Example usage
# Assuming you have a DataLoader for validation data (val_loader)
num_classes = 3  # Update with the correct number of classes in your classification problem
accuracy, f1, cm, auc_scores, class_report = evaluate_model(model, val_loader, device, num_classes)

print(f"Accuracy: {accuracy:.4f}")
print(f"F1-Score (Weighted): {f1:.4f}")
print("Confusion Matrix:")
print(cm)
print("AUC Scores:")
for class_label, auc_score in auc_scores.items():
    print(f"{class_label}: {auc_score:.4f}")
    
print("\nClassification Report:")
print(class_report)
```

```python
import torch
import matplotlib.pyplot as plt
import seaborn as sns
from transformers import DistilBertModel, DistilBertTokenizer

# Define the visualization function
def visualize_attention(model, input_text, tokenizer, device):
    # Tokenize the input text
    inputs = tokenizer(input_text, return_tensors="pt", padding=True, truncation=True).to(device)
    input_ids = inputs['input_ids']
    attention_mask = inputs['attention_mask']

    # Get embeddings and perform a forward pass through the model
    outputs = model(input_ids, attention_mask=attention_mask, output_attentions=True)  # Enable attentions
    last_hidden_state = outputs.last_hidden_state
    attention_weights = outputs.attentions

    # print(":::::::attention_weights::::", attention_weights)

    # Check if attention weights are available
    if attention_weights is not None and len(attention_weights) > 0:
        # Get attention weights from the first layer and first attention head
        attn_weights = attention_weights[0][0]  # Shape: [num_heads, seq_len, seq_len]

        # Convert attention weights to numpy and reshape for heatmap
        attn_weights = attn_weights.detach().cpu().numpy()

        # For visualization, we only care about the attention of the first head
        attn_weights_first_head = attn_weights[0]  # Shape: [seq_len, seq_len]

        # Plot the attention map for the first attention head
        plt.figure(figsize=(10, 8))
        sns.heatmap(attn_weights_first_head, cmap='viridis', xticklabels=input_text.split(), yticklabels=input_text.split(), annot=True, fmt='.2f')
        plt.title("Attention Weights - First Attention Head")
        plt.xlabel("Input Tokens")
        plt.ylabel("Input Tokens")
        plt.show()
    else:
        print("No attention weights captured")

# Example of how to use the function
def main():
    # Load pre-trained DistilBERT model and tokenizer
    model_name = "distilbert-base-uncased"
    model = DistilBertModel.from_pretrained(model_name)
    tokenizer = DistilBertTokenizer.from_pretrained(model_name)

    # Move model to device (GPU or CPU)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    # Example input text
    input_text = "This is a sample sentence for interpretability analysis."

    # Call the visualize_attention function
    visualize_attention(model, input_text, tokenizer, device)

if __name__ == "__main__":
    main()
```
